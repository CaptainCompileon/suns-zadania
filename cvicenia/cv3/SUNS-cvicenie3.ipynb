# Cvičenie č. 3

Na cvičení o 8 sme dostali dobrý postreh, stĺpec s názvom AdoptionSpeed (months) obsahuje hodnoty väčšie ako je vek zvieratka v stĺpci Age (months). Zrejme dostal nesprávny názov, mal by sa volať AdoptionSpeed (days), a teda vyjadruje počet dní po ktorých bolo zvieratko adoptované. Stĺpec si môžete premenovať, ďalej ho budeme uvádzať už s upraveným názvom.

1. Načítajte trénovacie a testovacie dáta , ktoré ste si pripravili na predošlom cvičení. Načítajte aj normalizéry a kódery.

2. Na tomto cvičení budete riešiť 2 úlohy:


*   Úloha č. 1 - klasifikácia, úlohou je predpovedať, ako rýchlo sa zvieratko adoptuje. Využite stĺpec AdoptionSpeed.
* Úloha č. 2 - regresia, úloha je rovnaká, tentokrát však bude výstupom číselná hodnota vyjadrujúca počet dní než sa zvieratko adoptovalo. Využite stĺpec AdoptionSpeed (days).



3. Vyberte si vhodnú podmnožinu vstupných parametrov pre jednotlivé úlohy. Sledujte závislosť atribútov od výstupných hodnôt jednotlivých úloh (pomôžte si korelačnou maticou). Vyhodnocujte pre trénovaciu množinu.
